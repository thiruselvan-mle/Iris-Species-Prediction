## Step 1: Import Libraries
We'll import libraries for visualization and data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Cleaned Dataset
Load the cleaned dataset

```python
df = pd.read_csv('D:\Thiru\ML_Projects\Iris-Species-Prediction\Data\Processed\Cleaned_iris.csv')
df.head()
```

## 3. Feature Distributions
Visualize the distribution of numeric features using histograms.

```python
df.hist(figsize=(10,8), bins=15, edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

## Step 4: Pairplot of Features
Use Seaborn pairplot to visualize relationships between features and the target variable.

```python
sns.pairplot(df, hue='Species', palette='Set2')
plt.show()
```

## Step 5: Correlation Analysis
Check the correlation between numeric features to identify strong relationships.

```python
corr = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
